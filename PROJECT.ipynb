# TASK 4

```python
import pandas as pd

class car_owner:
    def __init__(self, data):
        self.data = data

    def recommend(self, budget, min_range, min_battery):
        # Filter based on criteria
        filtered = self.data[
            (self.data['Minimal price (gross) [PLN]'] <= budget) &
            (self.data['Range (WLTP) [km]'] >= min_range) &
            (self.data['Battery capacity [kWh]'] >= min_battery)
        ]

        # Sort by range and battery capacity
        sorted_filtered = filtered.sort_values(
            by=['Range (WLTP) [km]', 'Battery capacity [kWh]'],
            ascending=False
        )

        # Select top 3 results
        top_evs = sorted_filtered[['Car full name', 'Minimal price (gross) [PLN]',
                                   'Range (WLTP) [km]', 'Battery capacity [kWh]']].head(3)

        return top_evs.reset_index()


# Load the Excel data
file_path = 'FEV-data.xlsx'  # Adjust the path if needed
excel_data = pd.ExcelFile(file_path)
df = excel_data.parse('Auta elektryczne')

# Create an instance of the recommender
owner = car_owner(df)

# Example usage
recommendations = owner.recommend(
    budget=400000,
    min_range=400,
    min_battery=80
)

print("Top 3 EV recommendations:")
display(recommendations)
```

# TASK 5

```python
from scipy.stats import ttest_ind

# Extracting engine power data for Tesla and Audi
tesla_power = fev[fev['Make'] == 'Tesla']['engine[km]'].dropna()
audi_power = fev[fev['Make'] == 'Audi']['engine[km]'].dropna()

# Calculating two-sample t-test
t_stat, p_val = ttest_ind(tesla_power, audi_power, equal_var=False)

# displaying result
print("=== Two-Sample T-Test: Tesla vs Audi (Engine Power) ===")
print(f"Tesla Mean Power: {tesla_power.mean():.2f} HP")
print(f"Audi Mean Power: {audi_power.mean():.2f} HP")
print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Tesla Sample Size: {len(tesla_power)}")
print(f"Audi Sample Size: {len(audi_power)}")

# Interpretation
if p_val < 0.05:
    print("\nConclusion: The difference in engine power is statistically significant.")
else:
    print("\nConclusion: No statistically significant difference in engine power between Tesla and Audi.")
```
